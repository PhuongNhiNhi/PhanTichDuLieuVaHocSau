```python
# Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 
# 12. 
#  Công thức tính: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
# df[tenconmoi] = df[]
```

```python
for i in ["1","2","6"]:
    print("TBM"+i)
```

```python
#(T*2 + L + H + S + V*2 + X + D + N) / 10
for i in ["1","2","6"]:
    df["TBM"+i] = (df["T"+i]*2 + df["L"+i]+ df["H"+i]+ df["S"+i]+ + df["V"+i]*2 + df["X"+i]+ df["D"+i]+ df["N"+i])/10
```

```python
df.rename(columns ={"TBM6":"TBM3"}, inplace =True)
```

```python
df.head()
```

```python
# Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 
# 10, 11, 12 như sau:
#  Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#  Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#  Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#  Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#  Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)
# df.loc[condition, "newcolumn"] = values_newcolumn
```

```python
for i in ["1","2","3"]:
    df.loc[df['TBM'+i] <= 5, 'XL'+i] = 'Y' 
    df.loc[(df['TBM'+i] <= 6.5) &(df['TBM'+i] > 5), 'XL'+i] = 'TB' 
    df.loc[(df['TBM'+i] < 8) &(df['TBM'+i] >= 6.5), 'XL'+i] = 'K' 
    df.loc[(df['TBM'+i] < 9) &(df['TBM'+i] >=8), 'XL'+i] = 'G' 
    df.loc[(df['TBM'+i] >= 9), 'XL'+i] = 'XS' 
```

```python
df.head()
```

```python
for i in ["1","2","3"]:
    df["US_TBM"+i] = df["TBM"+i]/10 *4
```

```python
# Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị ―1‖) và 
# rớt ( giá trị ―0‖) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
#  Với khối A, A1 nếu [(DH1*2 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, 
# ngược lại là rớt
#  Với khối B nếu [(DH1 + DH2*2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược 
# lại là rớt
#  Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược 
# lại là rớt

for i in df["KT"].unique():
    str(i)
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) < 5), "KQXT"] = 0
    
    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) < 5), "KQXT"] = 0
    
    df.loc[(i != "A1" or i != "A" or i == "B") & (((df["DH1"] + df["DH2"] + df["DH3"])/3) >= 5), "KQXT"] = 1
    df.loc[(i != "A1" or i != "A" or i == "B") & (((df["DH1"] + df["DH2"] + df["DH3"])/3) < 5), "KQXT"] = 0
```
